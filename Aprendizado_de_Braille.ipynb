````python
from gtts import gTTS
from IPython.display import Audio, display

def speak(text, lang='pt', filename='audio.mp3'):
    """
    Converte texto para fala usando gTTS e reproduz o áudio.
    Salva o áudio em um arquivo temporário e o reproduz no Colab.
    """
    try:
        tts = gTTS(text=text, lang=lang, slow=False)
        tts.save(filename)
        print(f"Reproduzindo: '{text}'")
        display(Audio(filename, autoplay=True))
    except Exception as e:
        print(f"Erro ao gerar ou reproduzir áudio: {e}")

# Exemplo de uso:
speak("Olá! Bem-vindo ao aprendizado de Braille.")
````

````python
import random
import speech_recognition as sr

global_score = 0 # Initialize global score

def listen_for_answer():
    """
    Ouve o microfone e tenta reconhecer a fala como uma resposta.
    Retorna a resposta reconhecida em maiúsculas ou None em caso de erro.
    """
    r = sr.Recognizer()
    with sr.Microphone() as source:
        speak("Por favor, diga a letra.", lang='pt', filename='audio_ask_letter.mp3')
        print("Diga a letra...")
        try:
            # Ajusta o ruído ambiente por 1 segundo
            r.adjust_for_ambient_noise(source, duration=1)
            audio = r.listen(source, timeout=5) # Ouve por no máximo 5 segundos
        except sr.WaitTimeoutError:
            print("Não detectei fala. Por favor, tente novamente.")
            speak("Não detectei fala. Por favor, tente novamente.", lang='pt', filename='audio_no_speech.mp3')
            return None

    try:
        # Tenta reconhecer a fala usando o Google Web Speech API
        spoken_text = r.recognize_google(audio, language="pt-BR").upper()
        print(f"Você disse: {spoken_text}")
        return spoken_text
    except sr.UnknownValueError:
        print("Não foi possível entender o áudio. Por favor, tente novamente.")
        speak("Não foi possível entender o áudio. Por favor, tente novamente.", lang='pt', filename='audio_unknown.mp3')
        return None
    except sr.RequestError as e:
        print(f"Não foi possível solicitar resultados do serviço de reconhecimento de fala; {e}")
        speak("Ocorreu um erro no serviço de reconhecimento de fala. Verifique sua conexão.", lang='pt', filename='audio_recognition_error.mp3')
        return None

def quiz_atualizado():
    """
    Executa uma rodada do quiz Braille interativo com feedback de áudio e atualiza a pontuação global.
    Returns: True if the answer was correct, False otherwise.
    """
    global braille
    global global_score

    letra = random.choice(list(braille.keys()))

    print(f"\nPontuação atual: {global_score}")
    print("Que letra é essa?")
    mostrar_pontos(letra, braille[letra]) # Passa a letra para mostrar_pontos e ela já lê em voz alta

    # Usa a entrada de voz
    resposta = None
    while resposta is None:
        resposta = listen_for_answer()
        if resposta is None:
            # If speech recognition fails, allow to try again
            continue

    if resposta == letra:
        print("✅ Correto!")
        speak("Correto!", lang='pt') # Feedback de áudio para resposta correta
        global_score += 1
        return True
    else:
        print(f"❌ Errado! Era {letra}")
        speak(f"Errado! Era {letra}", lang='pt') # Feedback de áudio para resposta incorreta
        return False

def start_braille_game(num_rounds=5):
    """
    Inicia um jogo Braille com múltiplas rodadas e pontuação.
    """
    global global_score
    global_score = 0 # Reset score for a new game
    speak("Iniciando o jogo Braille!", lang='pt')
    print("\n--- Jogo Braille Iniciado ---")

    for i in range(num_rounds):
        print(f"\n--- Rodada {i + 1}/{num_rounds} ---")
        quiz_atualizado()

    print("\n--- Jogo Concluído ---")
    speak(f"Sua pontuação final é {global_score} de {num_rounds}!", lang='pt')
    print(f"Sua pontuação final: {global_score} de {num_rounds}")

# Para iniciar o jogo, chame a função:
# start_braille_game(num_rounds=5) # Joga 5 rodadas
# ou para uma rodada única:
# quiz_atualizado()
````

### Dificuldade Adaptativa no Quiz

Para tornar o quiz mais inteligente e adaptativo, implementaremos a seguinte lógica:

1.  **Controle de Erros:** Manteremos um registro das letras que o usuário errou.
2.  **Priorização:** O quiz priorizará as letras que foram mais difíceis para o usuário nas rodadas anteriores.
3.  **Limiar de Domínio:** Uma letra será considerada 'dominada' se for respondida corretamente por um certo número de vezes consecutivas. Letras dominadas serão perguntadas com menos frequência para focar nas que precisam de mais prática.
4.  **Reinício:** As letras erradas retornarão ao pool de perguntas prioritárias.

Esta abordagem visa personalizar a experiência de aprendizado, focando nas áreas onde o usuário mais precisa de prática.

### Adicionando Contador de Acertos Consecutivos

Para aprimorar ainda mais a dificuldade adaptativa, vamos adicionar um contador de acertos consecutivos para cada letra. Uma letra só será considerada 'dominada' (e removida do pool de letras prioritárias) após ser respondida corretamente um número predefinido de vezes seguidas (por exemplo, 2 ou 3 vezes).

Se uma letra que já havia sido dominada for respondida incorretamente, ela será removida do conjunto de letras dominadas e seu contador de acertos consecutivos será zerado, voltando para o pool de prática.

````python
import random
# import speech_recognition as sr # Comentar ou remover esta linha se não for usar reconhecimento de fala

global_score = 0
# Dicionário para rastrear o número de vezes que uma letra foi respondida incorretamente
incorrectly_answered_letters = {}
# Conjunto de letras dominadas
mastered_letters = set()
# Dicionário para rastrear o número de acertos consecutivos por letra
consecutive_correct_answers = {}

def listen_for_answer():
    """
    Ouve o microfone e tenta reconhecer a fala como uma resposta.
    Retorna a resposta reconhecida em maiúsculas ou None em caso de erro.
    Temporariamente alterado para input de texto devido a problemas com microfone no Colab.
    """
    speak("Por favor, diga a letra.", lang='pt', filename='audio_ask_letter.mp3')
    print("Diga a letra...")

    try:
        spoken_text = input("Sua resposta (digite): ").upper()
        print(f"Você disse: {spoken_text}")
        return spoken_text
    except Exception as e:
        print(f"Ocorreu um erro ao ler a entrada: {e}")
        return None

def quiz_atualizado(difficulty_level=1):
    """
    Executa uma rodada do quiz Braille interativo com feedback de áudio e atualiza a pontuação global.
    Adapta a seleção de letras com base no desempenho do usuário.
    """
    global braille, global_score, incorrectly_answered_letters, mastered_letters, consecutive_correct_answers

    # Determinar o pool de letras para esta rodada
    # Priorizar letras com mais erros
    if incorrectly_answered_letters:
        # Criar uma lista de letras a serem priorizadas com base no número de erros
        # Cada erro aumenta a chance de ser escolhida
        prioritized_pool = []
        for letter, errors in incorrectly_answered_letters.items():
            prioritized_pool.extend([letter] * (errors * 2)) # Duplica a chance por erro

        # Adicionar o restante das letras não dominadas para garantir variedade
        remaining_letters = [l for l in braille.keys() if l not in mastered_letters and l not in incorrectly_answered_letters]
        if remaining_letters:
            prioritized_pool.extend(random.sample(remaining_letters, min(len(remaining_letters), 3))) # Adiciona algumas para variedade

        if prioritized_pool:
            letra = random.choice(prioritized_pool)
        else:
            # Fallback if somehow prioritized_pool is empty
            # If all letters are mastered, reset or congratulate
            if not braille.keys() - mastered_letters:
                speak("Parabéns! Você dominou todas as letras! Reiniciando o quiz.", lang='pt')
                mastered_letters.clear()
                incorrectly_answered_letters.clear()
                consecutive_correct_answers.clear()
                letra = random.choice(list(braille.keys()))
            else:
                letra = random.choice([l for l in braille.keys() if l not in mastered_letters])
    else:
        # Se não há erros, escolha de todas as letras não dominadas
        available_letters = [l for l in braille.keys() if l not in mastered_letters]
        if available_letters:
            letra = random.choice(available_letters)
        else:
            # Se todas as letras foram dominadas, reinicie ou parabenize
            speak("Parabéns! Você dominou todas as letras! Reiniciando o quiz.", lang='pt')
            mastered_letters.clear()
            incorrectly_answered_letters.clear()
            consecutive_correct_answers.clear()
            letra = random.choice(list(braille.keys()))

    print(f"\nPontuação atual: {global_score}")
    print("Que letra é essa?")
    mostrar_pontos(letra, braille[letra])

    resposta = None
    while resposta is None:
        resposta = listen_for_answer()
        if resposta is None:
            continue

    if resposta == letra:
        print("✅ Correto!")
        speak("Correto!", lang='pt')
        global_score += 1

        # Increment consecutive correct answers
        consecutive_correct_answers[letra] = consecutive_correct_answers.get(letra, 0) + 1

        # Define threshold for mastering a letter
        mastery_threshold = 2 # Requires 2 consecutive correct answers to master

        if consecutive_correct_answers[letra] >= mastery_threshold:
            mastered_letters.add(letra)
            # Optional: provide audio feedback when a letter is mastered
            # speak(f"Você dominou a letra {letra}!", lang='pt')

        # If answered correctly, remove from incorrectly_answered_letters
        if letra in incorrectly_answered_letters:
            del incorrectly_answered_letters[letra]


    else: # resposta != letra
        print(f"❌ Errado! Era {letra}")
        speak(f"Errado! Era {letra}", lang='pt')

        # Reset consecutive correct answers for this letter
        consecutive_correct_answers[letra] = 0

        # Register error
        incorrectly_answered_letters[letra] = incorrectly_answered_letters.get(letra, 0) + 1

        # If a letter was "mastered" but now answered incorrectly, remove it from mastered_letters
        if letra in mastered_letters:
            mastered_letters.remove(letra)

    print(f"Acertos consecutivos: {consecutive_correct_answers}")

def start_braille_game(num_rounds=5, difficulty_level=1):
    """
    Inicia um jogo Braille com múltiplas rodadas e pontuação, com dificuldade adaptativa.
    """
    global global_score, incorrectly_answered_letters, mastered_letters, consecutive_correct_answers
    global_score = 0
    incorrectly_answered_letters.clear()
    mastered_letters.clear()
    consecutive_correct_answers.clear() # Clear consecutive answers for a new game

    speak("Iniciando o jogo Braille adaptativo!", lang='pt')
    print("\n--- Jogo Braille Adaptativo Iniciado ---")

    for i in range(num_rounds):
        print(f"\n--- Rodada {i + 1}/{num_rounds} ---")
        quiz_atualizado(difficulty_level)

    print("\n--- Jogo Concluído ---")
    speak(f"Sua pontuação final é {global_score} de {num_rounds}!", lang='pt')
    print(f"Sua pontuação final: {global_score} de {num_rounds}")
    print(f"Letras com erros acumulados: {incorrectly_answered_letters}")
    print(f"Letras dominadas: {mastered_letters}")
    print(f"Acertos consecutivos finais: {consecutive_correct_answers}")
````
